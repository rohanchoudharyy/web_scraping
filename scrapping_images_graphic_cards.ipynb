```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = 'https://www.newegg.com/global/in-en/Desktop-Graphics-Cards/SubCategory/ID-48?Tid=1300354'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
print(soup.prettify())
```

```python
def download_pic(photo_url):
    name = photo_url.split('/')
    name = name[-2] + name[-1] #naming the pictures.
    urllib.request.urlretrieve(photo_url,name)

frames = soup.findAll('div',{'class','item-container'})
for i in frames:
    temp=i.find('a')
    photo_url=temp.img['src']
    photo_url=photo_url.replace('//','https://') 
    
    print(photo_url)
    download_pic(photo_url)
```
