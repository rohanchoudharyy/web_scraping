```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
response = requests.get(url)
response
```

```python
#whole html document of this website
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
table = soup.find('div',{'class':'rankings-block__container rankings-table'})
table
```

```python
rank = 1
for team in table.findAll('td',{'class':'table-body__cell rankings-table__team u-text-left'}):
    print(rank)
    print(team.text)
    rank=rank+1
    
```

```python
# Structuring the data according to our needs
rank = 1
for team in table.findAll('td',{'class':'table-body__cell rankings-table__team u-text-left'}): #specifying 'td' is important
    name = team.text.replace(" ","") #removing the spaces.
    name = team.text.replace("\n","") #removing the line spacings.
    
    print(str(rank) + name)
    rank=rank+1
    
```
