```python
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = 'https://stackoverflow.com/questions/16627227/http-error-403-in-python-3-web-scraping'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
print(soup.prettify())
```

```python
answers = soup.findAll('div',{'class':'answer'})
count=1
for i in answers:
    temp1 = i.find('div',{'class':'js-vote-count grid--cell fc-black-500 fs-title grid fd-column ai-center'})
    vote = temp1.text
    print("number of votes : ", vote)
    temp2 = i.find('div',{'class':'post-text'})
    solution = temp2.text

    print("Answer", count ," : ",  solution )
    count=count +1 
    
```
