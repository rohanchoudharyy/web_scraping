```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/Sachin_Tendulkar'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
#beautifulsoup will pass on the html content available in the response to the variable soup
text=soup.b.text
print(type(text))
print(text)
```

```python
all_data = soup.findAll('p')
```

```python
for i in all_data:
    print(i.text)
    
```
